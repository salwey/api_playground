## D & D

```python
def flatten(data, col_list, include_original_name=False):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        if include_original_name:
            flattened.columns = [column + '_' + colname for colname in columns]
        else:
            flattened.columns = [colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
```

```python
url = "https://api.open5e.com/monsters/"
response = requests.get(url)
response = response.json()
```

```python
dd_data = pd.DataFrame(response)
dd_data = dd_data[['results']]
dd_data = flatten(dd_data, ['results'])
dd_data
```
